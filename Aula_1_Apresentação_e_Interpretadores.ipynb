## Linha de comando 📝

Linha de comando ou interface de linha de comando, é uma aplicação de texto para ver e manipular arquivos em seu computador. É como o Windows Explorer ou o Finder no Mac, mas sem a interface gráfica. Outros nomes para a linha de comando são: cmd, CLI, prompt, console ou terminal.

### Windows

1.   Vá para o menu ou tela de iniciar, e digite "Prompt de comando" no campo de busca.
2.   Vá para o menu Iniciar → Sistema Windows → Prompt de comando
3.   Vá em Iniciar → Todos os Programas → Acessórios → Prompt de comando
4.   Vá para a tela Iniciar, passe o mouse sobre o canto inferior esquerdo da tela, e clique na seta para baixo que aparecerá (Em touch screen, passe o dedo sobre a parte inferior da tela). A página "Aplicativos" deve abrir. Clique em Prompt de comando na seção Sistema Windows.
5.   Pressione a tecla Windows no seu teclado e aperte a tecla "X" em seguida. Escolha "Prompt de comando" no menu pop-up.
6.   Mantenha pressionada a tecla Windows e pressione a tecla "R" para obter uma janela de "Executar". Digite "cmd" na caixa e clique no botão OK.


### OS X
1.   Vá para Aplicações → Utilidades → Terminal.

### Linux
1.   Provavelmente está em aplicações → Acessórios → Terminal, ou Aplicativos → Sistema → Terminal, mas isso pode depender do seu sistema. Se não estiver lá, você pode tentar utilizar o Google. :)

# Expressões aritméticas

Uma expressão aritmética é formada por números (chamados de operandos) e operadores.
Os operadores podem realizar soma (+), subtração (-), divisão (/) ou multiplicação (*) dos operandos.

## Precedência e associatividade

As regras de precedência indicam qual operador é calculado primeiro. Por exemplo, qual o resultado da expressão 2 - 3 * 4?

Como a multiplicação tem maior prioridade que a subtração, o produto 3*4 é reduzido ao valor 12 e a seguir se calcula o valor da subtração 2 - 12, resultando em -10.

As regras de associatividade indicam a ordem dos cálculos para operadores que tenham a mesma precedência.
